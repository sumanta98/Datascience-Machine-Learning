```python
import pandas as pd
import re
import numpy as np

class Bop():
    
    def readcsv(self):
        
        try:
            rd = pd.read_csv("bank-data.csv")
            unique_profession=(set(rd["job"]))
            
            age_min_max = rd.groupby(['job']).agg({'age':[min,max]})
            age_min_max.columns = ["_".join(x) for x in age_min_max.columns.ravel()]
            
            #print(age_min_max)
            
            return (unique_profession,age_min_max)
        except Exception as e:
            print('Error :',e)
        
            

    def eligible(self,unique_profession1,data2):
        
        try:
            print(data2)
            while True:
                prof= input('\nEnter your profession :\n')
                
                if unique_profession1 !='':
                    
                    if prof.lower() in unique_profession1:
                        print('\nclient is eligible\n')
                    elif prof.upper() =="END":
                        print('\nThank you a lot\n')
                        break
                    else:
                        print('\nclient is not eligible\n')
                else:
                    print('\nCSV file is empty')
        except Exception as e:
            print('Error',e)
        

obj1=Bop()
data,data2=obj1.readcsv()
obj1.eligible(data,data2)
```
