1. Plot Total Sales Per Month for Year 2011. 
</font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


read_bigmarker = pd.DataFrame(pd.read_csv('BigMartSalesData.csv'))

final = read_bigmarker.query('Year==2011').filter(['Month','Amount']).groupby(['Month'],as_index=False).sum()

x= final['Month']
y= final['Amount']


plt.plot(x,y)
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Amount Vs Month')
plt.show()
```

<font color='red'>How the total sales have increased
over months in Year 2011 ?</font>

Ans: 11th Month, it has increased and 12th month it has decrised again

<font color='red'>Which month has lowest Sales?</font>


Ans: As per plot , 2nd month has lowest Sales

3. Plot Pie Chart for Year 2011 Country Wise. 
</font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(12,6))
read_bigmarker = pd.DataFrame(pd.read_csv('BigMartSalesData.csv'))

final = read_bigmarker.query('Year==2011').filter(['Country','Amount']).groupby(['Country'],as_index=False).sum()

patch,text=plt.pie(final['Amount'],shadow=True, startangle=90)

plt.legend(patch,final['Country'],loc='Best')
plt.axis('equal')

plt.show()
```

<font  color='red'>
 Which Country contributes highest
towards sales?
</font>

Ans : United Kingdom

4. Plot Scatter Plot for the invoice amounts and see the concentration of amount.
In which range most of the invoice amounts are concentrated 

</font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(12,6))
read_bigmarker = pd.DataFrame(pd.read_csv('BigMartSalesData.csv'))

final = read_bigmarker.query('Year==2011').filter(['InvoiceDate','Amount']).groupby(['InvoiceDate'],as_index=False).sum()

x=final['InvoiceDate']
y=final['Amount']
plt.scatter(x,y,c='blue')
plt.xlabel('InvoiceDate')
plt.ylabel('Amount')
plt.title('Invice Vs Amount')
plt.show()
```
