<u>Business challenge/requirement</u><br>
You are a data analyst with University of Cal USA (Not a machine learning expert yet
as you still have not completed ML with Python Course :-)).The University has data of
Math, Physics and Data Structure score of sophomore students. This data is stored in
different files. The University has hired a data science company to do analysis of
scores and find if there is any correlation of score with age, ethnicity etc. Before the
data is given to the company you have to do data wrangling.<br>
<u>Key issues</u><br>
Ensure students identify is not revealed to the agency and only relevant data is
shared<br>
<u>Considerations</u><br>
NONE<br>
<u>Data volume</u><br>
-In thousands, but only around 1800 records are shared in files MathScoreTerm1.csv
DSScoreTerm1.csv, PhysicsScoreTerm1.csv<br>
Additional information<br>
-NA<br>
<u>Business benefits</u><br>
University can get more students enrollment by improving its international ranking
through personalized course/curriculum for students<br>
Approach to Solve<br><br>
You have to use fundamentals of Numpy and Pandas covered in module 4.
1. Read the three csv files which contains the score of same students in term1 for
each Subject
2. Remove the name and ethnicity column (to ensure confidentiality)
3. Fill missing score data with zero
4. Merge the three files
5. Change Sex(M/F) Column to 1/2 for further analysis
6. Store the data in new file – ScoreFinal.csv<br><br>
<u>Enhancements for code</u><br><br>
You can try these enhancements in code<br>

1.2. Fill the missing score for a student to the average of the class
    <br><br></font>

```python
import pandas as pd

#Read the three csv files which contains the score of same students in term1 for each Subject

read_math= pd.DataFrame(pd.read_csv('MathScoreTerm1.csv'))
read_phy = pd.DataFrame(pd.read_csv('PhysicsScoreTerm1.csv'))
read_DS  = pd.DataFrame(pd.read_csv('DSScoreTerm1.csv'))

print(read_math.head())
print(read_phy.head())
print(read_DS.head())


# Remove the name and ethnicity column (to ensure confidentiality)

del read_math['Name']
del read_math['Ethinicity']

del read_phy['Name']
del read_phy['Ethinicity']

del read_DS['Name']
del read_DS['Ethinicity']

print(read_math.head())
print(read_phy.head())
print(read_DS.head())

# Fill missing score data with zero
read_math.fillna(0)
read_phy.fillna(0)
read_DS.fillna(0)

print(read_math.head())
print(read_phy.head())
print(read_DS.head())

# Merge the three files

final = read_math.merge(read_DS,on='ID',how='inner').merge(read_phy,on='ID',how='inner').filter(['ID','Score_x','Score_y','Score','Age','Sex']).rename(columns={'Score_x':'Math_Score','Score_y':'DataStructure_Score','Score':'Physics_Score'})

print(final)
# Change Sex(M/F) Column to 1/2 for further analysis

final['Sex']=[1 if i=='M' else 2 for i in final['Sex']] 

print(final)

# Store the data in new file – ScoreFinal.csv

final.to_cs('ScoreFinal.csv')

```
