<font color='red'>1. You are given a dataset, which is present in the LMS, containing the number of
hurricanes occurring in the United States along the coast of the Atlantic. Load the
data from the dataset into your program and plot a Bar Graph of the data, taking
the Year as the x-axis and the number of hurricanes occurring as the Y-axis. <br><br>

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

read = pd.read_csv('Hurricanes.csv')

plt.bar(read['Year'],read['Hurricanes'])
plt.title('Hurricanes Statistics in United States along the coast of the Atlantic')
plt.xlabel('Year')
plt.ylabel('Hurricanes')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

dt={'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
'female': [0, 1, 1, 0, 1],
'age': [42, 52, 36, 24, 73],
'preTestScore': [4, 24, 31, 2, 3],
'postTestScore': [25, 94, 57, 62, 70]}

data =pd.DataFrame(dt,columns = ['first_name', 'last_name', 'age', 'female', 'preTestScore', 'postTestScore'])
print(data)
```

```python
x=data.preTestScore
y=data.postTestScore
plt.scatter(x,y,s=data.age)
plt.show()
```

```python
x=data.preTestScore
y=data.postTestScore
plt.scatter(x,y,s=300,c=data.female)
plt.show()
```
