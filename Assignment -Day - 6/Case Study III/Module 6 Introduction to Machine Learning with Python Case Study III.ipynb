```python
data

y= data['Yearly_Amount_Spent']
data.drop('Email',axis=1,inplace=True)
data.drop('Address',axis=1,inplace=True)
data.drop('Avatar',axis=1,inplace=True)  
x=data.iloc[:,0:4]
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=.33,random_state=85)
```

If we execute our code multiple times then the train and test dataset will be same in everytime for random_state=85.If We change ramdom state then the result will change again.

```python
lm = LinearRegression()

lm.fit(x_train,y_train)

y_pred = lm.predict(x_test)
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Actual Data')
plt.ylabel('Predicted Data')
plt.show()
```

```python
print(np.sqrt(metrics.mean_absolute_error(y_pred,y_test)))
```
